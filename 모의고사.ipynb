```python
#마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 인원을 가진 나이대는? (0~9 : 0 , 10~19 : 10)


df2=df.copy()
df2.age=df2.age//10 * 10
df2.groupby('age').size().sort_values(ascending=False).index[0]
```

```python
# balance 컬럼값들의 평균값 이상을 가지는 데이터를 ID값을 기준으로 내림차순 정렬했을때 상위 100개 데이터의 balance값의 평균은?

df3=df[df.balance >= df.balance.mean()].sort_values(by='ID',ascending=False).head(100)
df3.balance.mean()
```

```python
 #y 변수와 education 변수는 독립인지 카이제곱검정을 통해 확인하려한다. p-value값을 출력하라

v= pd.crosstab(df.y,df.education)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(v)
display(v)

print(p)
```

```python
# 각 job에 따라 divorced/married 인원의 비율을 확인 했을 때 그 값이 가장 높은 값은?


df2=df.groupby(['job','marital']).size().reset_index()
df3=df2.pivot_table(index='job',columns='marital')[0]
df3
```

```python
df3=df3.fillna(0)
df3['ratio']=df3.divorced/df3.married
df3.sort_values('ratio').ratio.values[-1]
```

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/test.csv')
submission= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/submission.csv')

display(train.head())
display(test.head())
display(submission.head())
```

```python
train.info()
train.nunique()
```

```python
submission.predict.value_counts()
```

```python
drop_col = ['ID','balance']
train.drop(columns=drop_col, inplace=True)
test.drop(columns=drop_col, inplace=True)
y=train.y
train.drop(columns='y',inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

train_dummies = pd.get_dummies(train)
test_dummies = pd.get_dummies(test)

x_train,x_test,y_train,_y_test = train_test_split(train_dummies,y,random_state=42)
```

```python
rf= RandomForestClassifier()
rf.fit(x_train,y_train)
pred=rf.predict_proba(x_test)
```

```python
from sklearn.metrics import roc_auc_score, classification_report
```

```python
test_pred= rf.predict_proba(test_dummies)[:,1]
len(test_pred)
```

```python
submission.predict = test_pred
```

```python
print('test roc score : ',roc_auc_score(_y_test,pred[:,1]))
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/pok/Pokemon.csv')
df.head()
```

```python
# 각 세대 중(Generation 컬럼)의 Speed와 Defense 컬럼의 차이(절댓값)이 가장 큰 세대는?


df.groupby('Generation')[['Speed','Defense']].mean().T.diff().T.sort_values('Defense').index[-1]
```

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/test.csv')

display(train.head())
display(test.head())
```

```python
train.info()
```

```python
y=train.stroke
```

```python
train.nunique()
```

```python
drop_col =['id','avg_glucose_level']
train.drop(columns=drop_col,inplace=True)
test.drop(columns=drop_col,inplace=True)
train.drop(columns='stroke',inplace=True)
train['age'] =train['age'].str.replace('*','').astype('int')

avg=train.bmi.mean()
train.bmi=train.bmi.fillna(avg)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

train_dummies = pd.get_dummies(train)
test_dummies = pd.get_dummies(test)
```

```python
x_train,x_test,y_train,y_test= train_test_split(train_dummies,y,stratify=y,test_size=0.3,random_state=23)

rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
pred= rf.predict_proba(x_test)

from sklearn.metrics import roc_auc_score,classification_report
roc_auc_score(y_test,pred[:,1])
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/spotify/spotify.csv')
df.head()
```

```python
# 데이터는 현재 년도별 100곡이 인기순으로 정렬되어 있다. 각 년도별 1~100위의 랭킹을 나타내는 rank컬럼을 만들고 매년도 1위의 bpm컬럼의 평균값을 구하여라
df=df.dropna()
df.loc[:,'rank'] = list(range(1,101))*10
df[df['rank']==1].bpm.mean()
```

```python
#2015년도에 가장많은 top100곡을 올린 artist는 누구인가?

df[df['top year']==2015].artist.value_counts().index[0]
```

```python
# 년도별 rank값이 1~10위 까지의 곡들 중 두번째로 많은 top genre는 무엇인rk
df[df['rank'].isin(range(1,11))]['top genre'].value_counts().index[2]
```

```python
# 피처링의 경우 title에 표시된다. 피처링을 가장 많이 해준 가수는 누구인가?

df[df.title.str.contains('feat.')].groupby('artist').size().sort_values(ascending=False)
```

```python
df.title.str.split('feat.').str[1].dropna().str[:-1].str.strip().value_counts()
```

```python
# top year 년도를 기준으로 발매일(year released)과 top100에 진입한 일자
 #(top year)가 다른 곡의 숫자를 count 했을때 가장 많은 년도는?
df[df['year released'] != df['top year']]['top year'].value_counts().index[0]
```

```python
len(df[df.artist.str.lower().str.contains('q')].artist.unique())
```

```python
abs(df.sort_values(by='rank',ascending=False)[:500].dur.mean() - df.sort_values(by='rank',ascending=False)[500:].dur.mean())
```

```python
# title을 띄어쓰기 단어로 구분 했을때 가장 많이 나온 단어는 무엇인가? (대소문자 구분 x)
df.title.str.split('\(feat').str[0].str.split().explode().str.lower().value_counts()

#df.title.str.lower().str.split("\(feat").value_counts()
```

```python
# 년도별 nrgy값의 평균값을 구할때 최대 평균값과 최소 평균값의 차이를 구하여라

df.groupby('top year').nrgy.mean().sort_values().values[9]-df.groupby('top year').nrgy.mean().sort_values().values[0]
```
