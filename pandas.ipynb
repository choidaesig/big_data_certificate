# 예제데이터 사용

```python
housing_df = pd.read_csv('/content/sample_data/california_housing_train.csv')
housing_df
```

```python
#info 함수 : Non-null 데이터 개수, 컬럼의 타입, 기초 통계 정보 조회
housing_df.info()
```

```python
#describe함수: 데이터 개수, 평균값, 표준편차 최솟값 사분위수값 최대값
housing_df.describe()
```

```python
# value_counts 해당 컬럼값의 유형과 개수 확인
house_mid = housing_df['housing_median_age']
val_cnt=house_mid.value_counts()
val_cnt
```

```python
housing_df.head(5)
```

```python
# iloc() // 행과열의 값으로 정수형 또는 정수형의 슬라이싱 값을 넣어야함 , iloc의 위치 기반 인덱싱
housing_df.iloc[0,0] #첫번째 행,열
```

```python
housing_df.iloc[0,2] # 첫번째 행, 세번째 열
```

```python
# loc() 명칭 기반 인덱싱
housing_df.loc[0:4,'housing_median_age']
```

## 집계함수

```python
#groupby
house_group = housing_df.groupby('housing_median_age').mean()
house_group.head(5)
```

```python
# agg
house_group = housing_df.groupby('housing_median_age')['total_rooms'].agg([min,max,sum])
house_group.head(5)
```

```python
# 결측치 처리하기
import numpy as np

housing_df['Age_na']=np.nan  # 결측치 확인
print(housing_df.isna().head(3))
```

```python
housing_df.isna().sum()
```

```python
housing_df['Age_na']=housing_df['Age_na'].fillna(housing_df['housing_median_age'].mean())
housing_df.head(3)
```
