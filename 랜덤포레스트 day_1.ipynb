```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
x_train.info()
```

```python
x_train.nunique()
```

```python
drop_col = ['CustomerId','Surname','Balance','EstimatedSalary']
x_train.drop(columns=drop_col,inplace=True)
x_test.drop(columns=drop_col,inplace=True)
```

```python
y=y_train.Exited
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
x_train_dummies = pd.get_dummies(x_train)
x_test_dummies = pd.get_dummies(x_test)
x_test_dummies = x_test_dummies[x_train_dummies.columns]
```

```python
X_train,X_validation,Y_train,Y_validation = train_test_split(x_train_dummies,y, test_size=0.3, random_state=11)

rf= RandomForestClassifier(random_state=11)
rf.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, roc_auc_score, precision_score
```

```python
predict_train_label = rf.predict(X_train)
predict_train_prob = rf.predict_proba(X_train)[:,1]

predict_valid_label = rf.predict(X_validation)
predict_valid_prob = rf.predict_proba(X_validation)[:,1]
```

```python
print ('accuracy',accuracy_score(Y_train,predict_train_label))
print ('accuracy',accuracy_score(Y_validation,predict_valid_label))
print('f1',f1_score(Y_train,predict_train_label))
print('f1',f1_score(Y_validation,predict_valid_label))
print('recall',recall_score(Y_train,predict_train_label))
print('recall',recall_score(Y_validation,predict_valid_label))
print('precision',precision_score(Y_train,predict_train_label))
print('precision',precision_score(Y_validation,predict_valid_label))
print('auc', roc_auc_score(Y_train,predict_train_prob))
print('auc',roc_auc_score(Y_validation,predict_valid_prob))
```

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
x_train.info()
x_train.nunique()
```

```python
drop_col =['StudentID']


x_test.drop(columns=drop_col, inplace=True)
x_train.drop(columns=drop_col, inplace=True)

y=y_train.G3
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import numpy as np

x_train_dummies = pd.get_dummies(x_train)
x_test_dummies = pd.get_dummies(x_test)
x_test_dummies = x_test_dummies[x_train_dummies.columns]

X_train,X_test,Y_train,Y_test=train_test_split(x_train_dummies,y)
rf=RandomForestRegressor(random_state=42,max_depth=8)
rf.fit(X_train,Y_train)

y_validation_pred = rf.predict(X_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score, mean_absolute_percentage_error
print('mse',mean_squared_error(Y_test,y_validation_pred))
print('mae',mean_absolute_error(Y_test,y_validation_pred))
print('mape',mean_absolute_percentage_error(Y_test,y_validation_pred))
print('rmse',np.sqrt(mean_absolute_percentage_error(Y_test,y_validation_pred)))
print('R2',r2_score(Y_test,y_validation_pred))
```
